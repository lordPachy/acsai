# Importing data

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

digits = load_digits()
print(digits.images.shape)
```

## Viewing some samples

```python
fig, axes = plt.subplots(5, 5, figsize=(8, 8), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace = 0.1, wspace = 0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap = 'binary')
    ax.text(0.05, 0.05, str(digits.target[i]), color='green')

plt.show()
```

## Splitting the data into training and testing dataset

```python
x = digits.data
x.shape
```

```python
y = digits.target
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, random_state=0, train_size = 0.8)
```

# Creating and training the model

```python
model = GaussianNB()
```

```python
model.fit(Xtrain, ytrain)
```

```python
y_pred = model.predict(Xtest)
print(accuracy_score(ytest, y_pred))
```

```python
mat = confusion_matrix(ytest, y_pred)

sns.heatmap(mat, square=True, annot=True, cbar=False)

plt.xlabel('predicted value')
plt.ylabel('true value')
```
